```python
import torch
import torch.nn as nn
import torch.nn.functional as F # Will allow us to move our data forward in our function

```

```python
# Create a model class that inherits nn.Module
class simple_model(nn.Module):
    # input layer(4 features of the flower)--> 
    # Hidden layer1-->
    # Hidden layer2-->
    # Output(3 classes of iris flower)
    
    def  __init__(self, input_features=4, h1=8, h2=9, output_features=3):
        super().__init__() # Instantiate our nn.Module
        self.fc1 = nn.Linear(input_features, h1)
        self.fc2 = nn.Linear(h1, h2)
        self.out = nn.Linear(h2, output_features)
    
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.out(x))
        
        return x
    
    
    
```

```python
# Pick a manual seed for randomization
torch.manual_seed(32)
# Create a instant model
model = simple_model()
```

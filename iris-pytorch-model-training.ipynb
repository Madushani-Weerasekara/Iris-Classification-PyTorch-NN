```python
import torch
import torch.nn as nn
import torch.nn.functional as F # Will allow us to move our data forward in our function

```

```python
# Create a model class that inherits nn.Module
class simple_model(nn.Module):
    # input layer(4 features of the flower)--> 
    # Hidden layer1-->
    # Hidden layer2-->
    # Output(3 classes of iris flower)
    
    def  __init__(self, input_features=4, h1=8, h2=9, output_features=3):
        super().__init__() # Instantiate our nn.Module
        self.fc1 = nn.Linear(input_features, h1)
        self.fc2 = nn.Linear(h1, h2)
        self.out = nn.Linear(h2, output_features)
    
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.out(x))
        
        return x
    
    
    
```

```python
# Pick a manual seed for randomization
torch.manual_seed(32)
# Create a instant model
model = simple_model()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import the iris dataset from scikit-learn's dataset module
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()
```

```python
# Trsin test split , set x,y
x = iris.data # features
y = iris.target # targets
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Train Test Split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=32)
```

```python
# Convert x features from numpy array to  float tensors
x_train_tensor = torch.FloatTensor(x_train)
x_test_tensor = torch.FloatTensor(x_test)
```

```python
# Convert y targets from numpy array to tensors long
y_train_tensor = torch.LongTensor(y_train)
y_test_tensor = torch.LongTensor(y_test)
```

```python
y_train_tensor
```

```python
y_test_tensor
```

```python
# set the crieterion of model to measure the error, how far the predictions from the dataset
crieterion = nn.CrossEntropyLoss()
# Choose Optimizer, lr=learning rate(If error doesn't go down after a bunch of itterations(epochs), lower our learning rate)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
model.parameters
```

```python
# Train the model
# epochs? (one round through all the training data in our network)
epochs = 100

losses = []

for i in range(epochs):
    # Go forward and get a prediction
    y_pred = model.forward(x_train_tensor) # get predicted results
    
    # Measure the error
    
    loss = crieterion(y_pred, y_train_tensor) # predicted value Vs. the y_trains
    
    # Keep trakc of our losses
    losses.append(loss.detach().numpy())
    
    # Print every 10 epochs
    if i % 10 == 0:
        print(f'Epochs: {i} and loss: {loss}')
        
    # Do some back propogation: take the error rate of forward propogation and feed it back through the network to fine tune the weights
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
# graph it out!
plt.plot(range(epochs), losses)
plt.ylabel("Loss / Error")
plt.xlabel("Epochs")
```
